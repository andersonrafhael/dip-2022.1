```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import cv2 as cv
```

### General Functions

```python
def plotImage(img):
    
    plt.figure()
    plt.subplots(figsize=(10, 7))
    
    
    plt.subplot(121)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)) 
    plt.title('Original Image')
    plt.xticks([]), plt.yticks([])


def showImages(original, result, tittle):
    
    plt.figure()
    plt.subplots(figsize=(10, 7))
    
    
    plt.subplot(121)
    plt.imshow(cv.cvtColor(original, cv.COLOR_BGR2RGB)) 
    plt.title('Original Image')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(122) 
    plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
    plt.title(tittle)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```

### Implement [image stitching](https://www.pyimagesearch.com/2018/12/17/image-stitching-with-opencv-and-python/) on images ['ic1.png', 'ic2.png', 'ic3.png', 'ic4.png'] to generate one panorama image 'ic.png'.

```python
stitcher = cv.Stitcher.create()
status, dst = stitcher.stitch([cv.imread("images/ic1.jpg"), cv.imread("images/ic2.jpg"), cv.imread("images/ic3.jpg"), cv.imread("images/ic4.jpg")])

plt.imshow(cv.cvtColor(dst, cv.COLOR_BGR2RGB)) 
plt.show()
```
