Processamento de Imagens Coloridas

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import cv2 as cv
```

```python
# General Functions

from cv2 import CMP_EQ


def showChannels(convertedImage, tittle):
    (C1, C2, C3) = cv.split(convertedImage)
    
    plt.figure()
    plt.subplots(figsize=(7, 5))
    
    
    plt.subplot(141)
    plt.imshow(convertedImage, cmap='gray') 
    plt.title(tittle)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(142)
    plt.imshow(C1, cmap='gray')
    plt.title('Channel 1')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(143) 
    plt.imshow(C2, cmap='gray')
    plt.title('Channel 2')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(144) 
    plt.imshow(C3, cmap='gray')
    plt.title('Channel 3')
    plt.xticks([]), plt.yticks([])
    
    plt.show()
    

def BGR2CMYK(original):
    
    # Extract channels
    img = original.astype(np.float64) / 255.
    K = 1 - np.max(img, axis=2)
    C = (1-img[..., 2] - K) / (1-K)
    M = (1-img[..., 1] - K) / (1-K)
    Y = (1-img[..., 0] - K) / (1-K)

    # Convert the input BGR image to CMYK colorspace
    dst = (np.dstack((C, M, Y, K)) * 255).astype(np.uint8)
    
    return dst
    
    
def showColorSpaces(original):
    
    NTSCImage = cv.cvtColor(original, cv.COLOR_BGR2YCrCb)
    HSVImage = cv.cvtColor(original, cv.COLOR_BGR2HSV)
    GRAYImage = cv.cvtColor(original, cv.COLOR_BGR2GRAY)
    LABImage = cv.cvtColor(original, cv.COLOR_BGR2LAB)
    CIEXYZImage = cv.cvtColor(original, cv.COLOR_BGR2XYZ)
    CMYKImage = BGR2CMYK(original)
    
    showChannels(NTSCImage, 'NTSC')
    showChannels(HSVImage, 'HSV')
    showImages(original, GRAYImage, 'Gray')
    showChannels(LABImage, 'LAB')
    showChannels(CIEXYZImage, 'CIEXYZ')
    showImages(original, CMYKImage, 'CMYK')
   
    

def showImages(original, result, tittle):
    
    plt.figure()
    plt.subplots(figsize=(5, 3))
    
    
    plt.subplot(121)
    plt.imshow(cv.cvtColor(original, cv.COLOR_BGR2RGB)) 
    plt.title('Original Image')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(122) 
    plt.imshow(result, cmap='gray')
    plt.title(tittle)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```

```python
img = cv.imread('img/baboon.png')
(B, G, R) = cv.split(img)

showImages(img, B, "Blue Channel")
showImages(img, G, "Green Channel")
showImages(img, R, "Red Channel")
cv.waitKey(0)
```
